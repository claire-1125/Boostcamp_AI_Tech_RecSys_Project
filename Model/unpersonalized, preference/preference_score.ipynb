```python
import numpy as np
from collections import defaultdict
```

```python
logs = [['2019-11-30 02:35:26 PM', 3, 6881099, 0, 4],
        ['2019-12-01 03:55:12 PM', 3, 6978911, 0, 5],
        ['2020-01-05 01:03:49 PM', 3, 6974675, 1],
        ['2020-01-06 08:27:03 PM', 3, 6974675, 0, 5]]
```

```python
"""
* 맨 마지막에 시간순으로 정렬할 필요가 없으면 시간 정보 필요 없을 것 같네요.....

예시) 
3번 user에 대한 레시피 로그 
- 19/11/30에 차돌박이 찜 레시피에 4점 부여
- 19/12/01에 참치마요덮밥 레시피에 5점 부여
- 20/01/05에 초콜릿쿠키 레시피를 내보내기
- 20/01/06에 초콜릿쿠키 레시피에 5점 부여


순서대로 timestamp, uid, rid, 구분, 로그값(평점)
logs = [[2019-11-30 02:35:26 PM, 3, 6881099, 0, 4],
        [2019-12-01 03:55:12 PM, 3, 6978911, 0, 5],
        [2020-01-05 01:03:49 PM, 3, 6974675, 1],
        [2020-01-06 08:27:03 PM, 3, 6974675, 0, 5]
]

순서대로 rid : [횟수, 계수, 선호도 값]
recipe_info = {6881099 : [1, 2, ],
        6978911 : [1, 3, ],
        6974675 : [2, 4, ]
}
"""

NUM_OF_EVENTS = len(logs)

# 각 이벤트에 대한 중요도
STAR_FIVE = 3
STAR_FOUR = 2
EXPORT = 1

# 선호도 threshold
# THRESHOLD = 0.5

# 유저 아이디
UID = logs[0][1]

recipe_info = defaultdict(list)
for event in logs:
    recipe_info[event[2]] = [0]*3


# 각 레시피 횟수 카운팅
for event in logs:
    recipe_info[event[2]][0] += 1


# 각 레시피에 대한 계수 계산하기
for rid in recipe_info:
    # 조건에 해당하는 element를 가지는 iterator 객체
    for event in filter(lambda x: x[2] == rid, logs):
        # 이벤트 구분 (0:별점, 1:내보내기)
        if event[3] == 0 and event[4] == 5:
            recipe_info[rid][1] += STAR_FIVE
        elif event[3] == 0 and event[4] == 4:
            recipe_info[rid][1] += STAR_FOUR
        elif event[3] == 1:
            recipe_info[rid][1] += EXPORT
  

# 각 레시피에 대한 선호도 계산
for rid in recipe_info:
    preference = np.tanh(recipe_info[rid][1] * (recipe_info[rid][0] / NUM_OF_EVENTS))
    recipe_info[rid][2] = preference


# # 선호도 값이 0.5 이상인 레시피만 따로 저장
# candidate = []
# for rid in recipe_info:
#     if rid[2] >= THRESHOLD:
#         candidate.append(rid)


# # 레시피 후보들을 시간순으로 정렬 (시간 자료형을 모르는구나...)

print(recipe_info)

"""
[
    (6881099, [1, 2, 0.46211715726000974]),
    (6978911, [1, 3, 0.6351489523872873]),
    (6974675, [2, 4, 0.9640275800758169]})
]
"""


# 가장 선호도가 높은 레시피 아이디를 추출
sorted_list = sorted(recipe_info.items(), key=lambda recipe:recipe[1][2], reverse=True)
print(sorted_list)

# return UID, sorted_dict[0][0]
```

```python
"""
* 맨 마지막에 시간순으로 정렬할 필요가 없으면 시간 정보 없어도 됩니다......ㅠㅠ


예시) 
3번 user에 대한 레시피 로그 
- 19/11/30에 차돌박이 찜 레시피에 4점 부여
- 19/12/01에 참치마요덮밥 레시피에 5점 부여
- 20/01/05에 초콜릿쿠키 레시피를 내보내기
- 20/01/06에 초콜릿쿠키 레시피에 5점 부여


순서대로 timestamp, uid, rid, 구분, 로그값(평점)
logs = [[2019-11-30 02:35:26 PM, 3, 6881099, 0, 4],
        [2019-12-01 03:55:12 PM, 3, 6978911, 0, 5],
        [2020-01-05 01:03:49 PM, 3, 6974675, 1],
        [2020-01-06 08:27:03 PM, 3, 6974675, 0, 5]
]

순서대로 rid : [횟수, 계수, 선호도 값]
recipe_info = {6881099: [1, 2, 0.46211715726000974], 
               6978911: [1, 3, 0.6351489523872873], 
               6974675: [2, 4, 0.9640275800758169]
}

sorted_list = [(6974675, [2, 4, 0.9640275800758169]), 
               (6978911, [1, 3, 0.6351489523872873]), 
               (6881099, [1, 2, 0.46211715726000974])
]
"""

def calculate_preference(logs: list) -> tuple:

    # 각 이벤트에 대한 중요도
    STAR_FIVE = 3
    STAR_FOUR = 2
    EXPORT = 1

    # 선호도 threshold
    # THRESHOLD = 0.5

    # 유저 아이디
    UID = logs[0][1]

    
    recipe_info = defaultdict(list)  # 각 레시피별 정보 담는 딕셔너리
    for event in logs:
        recipe_info[event[2]] = [0]*3


    ## 각 레시피 횟수 카운팅
    for event in logs:
        recipe_info[event[2]][0] += 1


    ## 각 레시피에 대한 계수 계산하기
    for rid in recipe_info:
        for event in filter(lambda x: x[2] == rid, logs):
            if event[3] == 0 and event[4] == 5:  # 별점 5점인 이벤트면 계수가 3
                recipe_info[rid][1] += STAR_FIVE
            elif event[3] == 0 and event[4] == 4:  # 별점 4점인 이벤트면 계수가 2
                recipe_info[rid][1] += STAR_FOUR
            elif event[3] == 1:  # 내보내기 이벤트면 계수가 1
                recipe_info[rid][1] += EXPORT
    

    ## 각 레시피에 대한 선호도 계산
    for rid in recipe_info:
        preference = np.tanh(recipe_info[rid][1] * (recipe_info[rid][0] / len(logs)))
        recipe_info[rid][2] = preference


    # ## 선호도 값이 0.5 이상인 레시피만 따로 저장
    # candidate = []
    # for rid in recipe_info:
    #     if rid[2] >= THRESHOLD:
    #         candidate.append(rid)


    # ## 레시피 후보들을 시간순으로 정렬 (시간 자료형을 모르는구나...)


    ## 가장 선호도가 높은 레시피 아이디를 추출
    sorted_list = sorted(recipe_info.items(), key=lambda recipe:recipe[1][2], reverse=True)

    return UID, sorted_list[0][0]
```

```python
uid, rid = calculate_preference(logs)
print("{}번 유저가 가장 좋아하는 레시피는 {}".format(uid,rid))
```
